```python
import os

malawi_folder = 'Malawi'
data_folder = 'data'

# Get a list of all TIFF files in the 'malawi' folder
tif_files = [filename for filename in os.listdir(malawi_folder) if filename.endswith('.tif')]

# Initialize a counter for deleted files
deleted_files_counter = 0

# Iterate over each TIFF file
for tif_file in tif_files:
    # Extract the filename without the extension
    shapefile_name = os.path.splitext(tif_file)[0]
    
    # Define the paths to the shapefile files
    # shp_file = os.path.join(data_folder, shapefile_name + '.shp')
    shx_file = os.path.join(data_folder, shapefile_name + '.shx')
    dbf_file = os.path.join(data_folder, shapefile_name + '.dbf')
    prj_file = os.path.join(data_folder, shapefile_name + '.prj')

    # Check if the shapefile files exist and delete them if found
    if os.path.exists(dbf_file):
        try:
            # os.remove(shp_file)
            os.remove(shx_file)
            os.remove(dbf_file)
            os.remove(prj_file)
            print(f"Deleted shapefile files for: {shapefile_name}")
            deleted_files_counter += 1
        except: 
            print("File not found...moving on....")
            continue

# Print the total number of deleted files
print(f"Deleted {deleted_files_counter} shapefile(s) in total.")
print("Process completed.")
```
