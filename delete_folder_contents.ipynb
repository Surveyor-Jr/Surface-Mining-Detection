```python
import os
import shutil

def delete_folder_contents(folder_path):
    # Check if the folder exists
    if not os.path.exists(folder_path):
        print(f"The folder '{folder_path}' does not exist.")
        return
    
    # Iterate over all the files and folders in the specified folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        
        # Check if it is a file
        if os.path.isfile(file_path):
            # Delete the file
            os.remove(file_path)
            print(f"Deleted file: {file_path}")
        # Check if it is a subfolder
        elif os.path.isdir(file_path):
            # Delete the subfolder and its contents recursively
            shutil.rmtree(file_path)
            print(f"Deleted folder: {file_path}")

# Specify the folder path to delete its contents
folder_path = "data"

# Call the function to delete the folder contents
delete_folder_contents(folder_path)
```
